## Imports

```python
import pandas as pd
import numpy as np
import sys
sys.path.append("../src")
from features import smiles_to_morgan
```

## Load Dataset

```python
df = pd.read_csv("../data/raw/tox21.csv.gz")
print(df.shape)
```

## Compute Morgan Fingerprints

```python
from tqdm import tqdm
tqdm.pandas()

print("Computing fingerprints...")
X = np.vstack(df['smiles'].progress_apply(smiles_to_morgan).values)
print(f"X shape: {X.shape}")
```

## Extract Target Matrix

```python
targets = ['NR-AR', 'NR-AR-LBD', 'NR-AhR', 'NR-Aromatase', 'NR-ER', 
           'NR-ER-LBD', 'NR-PPAR-gamma', 'SR-ARE', 'SR-ATAD5', 
           'SR-HSE', 'SR-MMP', 'SR-p53']

y = df[targets].values
print(f"y shape: {y.shape}")
```

## Save Processed Data

```python
np.save("../data/processed/X_morgan.npy", X)
np.save("../data/processed/y_targets.npy", y)
np.save("../data/processed/smiles.npy", df['smiles'].values)
print("Saved")
```
